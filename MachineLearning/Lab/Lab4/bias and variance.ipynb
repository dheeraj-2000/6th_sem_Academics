```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import random
```

```python
x = np.linspace(-1,1,20)
y = np.sin(3.14*x)
xn = np.linspace(-1,1,2)
plt.plot(x,y)
```

```python
def calculate(x1,p):
    rx = []
    for i in x1:
        cx = []
        for j in range(p):
            cx.append(i**j)
        rx.append(cx)
    return pd.DataFrame(rx)
```

```python
def linear_regression(df,test_y,test_x,p):
    w = np.matmul(np.linalg.inv(np.matmul(np.transpose(df.values),df.values)),np.matmul(np.transpose(df.values),test_y))

    df1 = calculate(xn,p)
    plt.plot(xn,np.matmul(df1.values,w),color = "green")
```

```python
def main(x,y,p):
    for i in range(len(x)):
        for j in range(i+1,len(x)):
            test_x = [x[i],x[j]]
            test_y = [y[i],y[j]]
            df = calculate(test_x,p)
           
            linear_regression(df,test_y,test_x,p)       
```

```python
main(x,y,1)
plt.plot(x,y,color = "black")
```
