```python
x = np.linspace(0,10,100)
y = 2*x + 3
w0 = np.linspace(-3,9,100)
w1 = np.linspace(-5,8,100)
```

```python
def like(x,y ,w0 , w1):
    ll=0 
    for i ,j in zip(x,y):
        ll+= np.log(norm.pdf(ww0*i+ww1,j,1))
    return ll
```

```python
ww0,ww1= np.meshgrid(w0,w1)
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.plot_surface(ww0, ww1, like(x,y,ww0,ww1))
```
