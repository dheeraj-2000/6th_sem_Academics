```python
import numpy as np   
import matplotlib.pyplot as plt 
from mpl_toolkits import mplot3d
```

```python
x = np.linspace(1, 11,10)
y = 5*x +1 + np.random.normal(0, 3, 100)
```

```python
plt.scatter(x, y)
```

```python
def erro(a1, a2, x, y):
    e = 0
    for i in range(len(x)):
        e += ((a1*x[i] - a0) - y[i] )**2
    return e
```

```python
a0 = np.linspace(-100,100,100)
a1 = np.linspace(-100,100,100) 
A0,A1 = np.meshgrid(a0,a1)
```

```python

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
mycmap = plt.get_cmap('gist_earth')
ax.plot_surface(a1, a0, erro(A1, A0, x, y),cmap = mycmap  )
```

```python
plt.contour(a1, a0, erro(A1, A0, x, y))
```
