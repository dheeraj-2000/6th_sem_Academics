

### Task 1: Pdf for one value in x and mean distributed in 100 parts from 0-10



```python
x = 5
m = np.linspace(0,10,100)
Y_ax = norm.pdf(x, m, 1 )
```

```python
plt.plot(m, Y_ax)
plt.xlabel('mean')
plt.ylabel('PDF')
```

```python
x = np.linspace(0, 10,100)
y = 5*x + np.random.normal(0, 3, 100)
w0 = np.linspace(-10,10,100)
w1 = np.linspace(-10,10,100)
```

```python
def like(x,y ,w0 , w1):
    ll=0 
    for i ,j in zip(x,y):
        ll+= np.log(norm.pdf(ww0*i+ww1,j,1))
    return ll
```

```python
ww0,ww1= np.meshgrid(w0,w1)
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
mycmap = plt.get_cmap('gist_earth')
ax.plot_surface(ww0, ww1, like(x,y,ww0,ww1), cmap = mycmap)
```

```python
plt.contour(ww0, ww1, like(x,y,ww0,ww1))
```
