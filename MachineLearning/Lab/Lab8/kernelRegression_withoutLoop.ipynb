```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
x = np.linspace(0,2*np.pi,100)
y = np.sin(x)
```

```python
x_data = np.array([0.1,0.2,1.25,1.4,1.45,2.0,3.0,4.0,5.0,6.6]).reshape(10,1)
y_data = np.sin(x_data)+(0.31*np.random.randn(10,1))
```

```python
def kernelCp(xp,x_datatemp):
    xtemp = np.asarray(xp)
    xtemp = xtemp.reshape(100,1)
    x_datatemp = x_datatemp.reshape(10,)
    x_datptemp = norm.pdf(xtemp,x_datatemp,1/2)   
    return x_datptemp
temp = kernelCp(x,x_data)
```

```python
sumval = np.sum(temp, axis=1)
val = temp.T/sumval
ktemp = val.T
```

```python
y_temporary = np.dot(ktemp,y_data)
```

```python
sd = np.std(y_data)
```

```python
fig = plt.figure()
plt.plot(x,y,color="lime")
plt.plot(x,y_temporary,color="r")
plt.scatter(x_data,y_data)
circle = list()
for i in range(0,10):
    circle.append(plt.Circle((x_data[i], y_data[i]), sd/5, color = 'blue', fill = False))
ax = plt.gca()
for i in range(0,10):
    ax.add_artist(circle[i])
```
