```python
import matplotlib.pyplot as plt
import networkx as nx
```

```python
def set_all_B(g):
    for each in g.nodes():
        g.nodes[each]['action'] = 'B'
```

```python
def set_A(g, list1):
    for each in list1:
        g.nodes[each]['action'] ='A'
```

```python
def get_colors(g):
    list1=[]
    for each in g.nodes():
        if g.nodes[each]['action']=='B':
            list1.append('red')
        else:
            list1.append('green')
    return list1
```
