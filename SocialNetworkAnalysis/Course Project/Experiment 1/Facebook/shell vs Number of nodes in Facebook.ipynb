```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("/home/dheeraj/my_projects/my_project_env/practice/6th_sem_Academics/SocialNetworkAnalysis/Course Project/Experiment 1/Facebook/shellvsNodeC.csv",encoding = "ISO-8859-1")
df
```

```python
df1= pd.DataFrame(df)
df1['Number_of_Nodes'] = df1["NodeCount"]
```

```python


df1.plot( x='shell', y='Number_of_Nodes', kind='bar',figsize=(18, 8), edgecolor='k', color= "lightgreen")

plt.xlabel('Shell Number', fontsize=20)
plt.ylabel('Number of Nodes', fontsize=20)
plt.title('Shell number vs Number of Nodes in Facebook friendship network', fontsize=20,fontweight='bold')
SMALL_SIZE = 8
MEDIUM_SIZE = 10
BIGGER_SIZE = 12

plt.rc('xtick', labelsize=BIGGER_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=BIGGER_SIZE)
```
